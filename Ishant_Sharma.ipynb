```python
import pandas as pd
```

```python
df = pd.read_excel('Ishant.xlsx')
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Assuming 'Year' is a datetime column, if not, you may need to convert it to a datetime type.
# You can use df['Year'] = pd.to_datetime(df['Year'], format='%Y') if needed.

plt.figure(figsize=(12, 6))  # Adjust the figure size
plt.plot(df['Year'], df['Wkts'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Avg')
plt.title('Ishant Sharma Wickets Over the Years')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  # Adjust the figure size

# Plotting Average (Avg)
plt.plot(df['Year'], df['Avg'], label='Average', marker='o', linestyle='-')

# Plotting Wickets (Wkts)
plt.plot(df['Year'], df['Wkts'], label='Wickets', marker='s', linestyle='-')

# Plotting Strike Rate (SR)
plt.plot(df['Year'], df['SR'], label='Strike Rate', marker='^', linestyle='-')

# Plotting Innings
plt.plot(df['Year'], df['Inns'], label='Innings', marker='^', linestyle='-')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Ishant Sharma Bowling Stats Over the Years')
plt.grid(True)
plt.xticks(df['Year'], rotation=45)

# Adding a legend to differentiate the lines
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'Year' is a datetime column, if not, convert it to datetime as mentioned earlier.
# You can use df['Year'] = pd.to_datetime(df['Year'], format='%Y') if needed.

# Find the most recent year in the dataset
most_recent_year = df['Year'].max()

# Calculate the year 5 years ago
year_5_years_ago = most_recent_year - pd.DateOffset(years=5)

# Filter the DataFrame to include only data from the last 5 years
last_5_years_df = df[df['Year'] >= year_5_years_ago]

# Adjust the figure size
plt.figure(figsize=(10, 6))  # You can change the width and height as per your preference

# Plotting Average (Avg)
plt.plot(last_5_years_df['Year'], last_5_years_df['Avg'], label='Average', marker='o', linestyle='-')

# Plotting Wickets (Wkts)
plt.plot(last_5_years_df['Year'], last_5_years_df['Wkts'], label='Wickets', marker='s', linestyle='-')

# Plotting Strike Rate (SR)
plt.plot(last_5_years_df['Year'], last_5_years_df['SR'], label='Strike Rate', marker='^', linestyle='-')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Ishant Sharma Bowling Stats Over the Last 5 Years')
plt.grid(True)

# Customize the x-axis to display all years
plt.xticks(last_5_years_df['Year'], rotation=45)

# Adding a legend to differentiate the lines
plt.legend()

plt.show()
```
